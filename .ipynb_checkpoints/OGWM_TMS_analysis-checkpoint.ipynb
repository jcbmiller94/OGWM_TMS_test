```python
import numpy as np
np.set_printoptions(precision = 3)
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
pd.set_option('precision',2)
import scipy.io as io
import json
from copy import deepcopy
import scipy.stats as stats
import seaborn as sns
#import pyvttbl as pt
#sns.reset_orig() # to reset to matplotlib defaults if need be 
# choose colors that work for most color-blind folks
#sns.set_palette("colorblind")
sns.set(color_codes=True)
sns.set_style('white')
```

```python
subjects = ['OGWM_101', 'OGWM_102', 'OGWM_104']

files = ['OGWM_retrocue_task_v05_OGWM_101_sess1_s1_APPENDED.txt', 
        'OGWM_retrocue_task_v05_OGWM_101_sess2_pfc_APPENDED.txt',
        'OGWM_retrocue_task_v05_OGWM_102_sess1_s1.txt',
        'OGWM_retrocue_task_v05_OGWM_102_sess2_pfc.txt',
        'OGWM_retrocue_task_v05_OGWM_104_sess1_pfc_APPENDED.txt',
        'OGWM_retrocue_task_v05_OGWM_104_sess2_s1_APPENDED.txt']
```

```python
# loop through and groupby compatibility to make new dataframe

df_combined = pd.DataFrame() # empty DataFrame 

for file in files:
    
    fname = file    

    df = pd.read_csv(fname, header = 0, sep = '\s+') # read in df for the subject
    df = df[['subject','session','target','block','TrialType','CompType', 'move_init_bound_msecRT', 'enter_box_msecRT', 'velocity_drop_msecRT', 'msecRT', 'ACC', 'cueValidity', 'probeResp', 'probemsecRT', 'probeACC']]
    df['msecRT'] = pd.to_numeric(df['msecRT'], errors = 'coerce')
    df['move_init_bound_msecRT'] = pd.to_numeric(df['move_init_bound_msecRT'], errors = 'coerce')
    df['enter_box_msecRT'] = pd.to_numeric(df['enter_box_msecRT'], errors = 'coerce')
    df['velocity_drop_msecRT'] = pd.to_numeric(df['velocity_drop_msecRT'], errors = 'coerce')
    df['probemsecRT'] = pd.to_numeric(df['probemsecRT'], errors = 'coerce')
    df_combined = pd.concat([df_combined, df])
```

```python
df_combined.sample(15)
```

### Report subject accuracies for motor and probe responses

```python
grouped_subject = df_combined.groupby('subject') # group by subject name 
grouped_subject = df_combined.groupby(['subject', 'target', 'TrialType'], as_index=False)
```

```python
df = grouped_subject[('ACC', 'probeACC')].aggregate(np.mean) # aggregate within group by passing np.mean function
df
```

```python
# only include first 6 blocks to make even for each subject 
df_combined = df_combined[(df_combined.block != 7)]
```

```python
# add a column to distringuish first vs second half of task 

half_bool = np.array(df_combined.block < 4)
half = np.empty((half_bool.shape), dtype = 'S15')

for i in range(half.shape[0]):
    if half_bool[i] == True:
        half[i] = 'First'
    else: 
        half[i] = 'Second'
        
df_combined['half'] = half
df_combined.sample(7)
```

```python
# plot settings

fs = 16
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
g = sns.factorplot(x="TrialType", y="ACC", hue="subject", col="target", data=df_combined, 
                   capsize=0.05, size=6, aspect=.75)
```

```python
# look at group mean 
g = sns.factorplot(x="TrialType", y="ACC", hue="target", data=df_combined, 
                   capsize=0.05, size=6, aspect=.75, order = ['compatible', 'neutral', 'incompatible'],
                 errwidth = 2.25)

g.set_xticklabels(['compatible','neutral','incompatible'], fontsize = fs + 2)
g.set_xlabels('\nTrial Type', fontsize = fs + 2)
g.set_ylabels('Accuracy', fontsize = fs + 2)
g.set_yticklabels(fontsize = fs)
sns.despine(right = False, top = False)

# look at group mean split by half 
# g = sns.factorplot(x="TrialType", y="ACC", hue="target", col = 'half', data=df_combined, 
#                    capsize=0.05, size=6, aspect=.75, order = ['compatible', 'neutral', 'incompatible'],
#                  errwidth = 2.25)
```

```python
# look at group mean 
g = sns.factorplot(x="TrialType", y="msecRT", hue="target", data=df_combined, 
                   capsize=0.05, size=6, aspect=.75, order = ['compatible', 'neutral', 'incompatible'],
                 errwidth = 2.25)

g.set_xticklabels(['compatible','neutral','incompatible'], fontsize = fs + 2)
g.set_xlabels('\nTrial Type', fontsize = fs + 2)
g.set_ylabels('Total RT (ms)', fontsize = fs + 2)
g.set_yticklabels(fontsize = fs)
sns.despine(right = False, top = False)


# look at group mean split by half 
g = sns.factorplot(x="TrialType", y="msecRT", hue="target", col = 'half', data=df_combined, 
                   capsize=0.05, size=6, aspect=.75, order = ['compatible', 'neutral', 'incompatible'],
                 errwidth = 2.25)
```
